```python
table_ratings = pd.DataFrame([df[df.Type == 0].Ratings]).T.reset_index(drop=True) #df.loc[df.Type == 1].Novelty]
table_ratings.rename(columns = {"Ratings": 'Typical'}, inplace = True)
for i, n in zip(range(1,3), ['Atyplical', "Ripple"]):
    column = pd.DataFrame([df.loc[df.Type == i].Ratings]).T.reset_index(drop=True)
    column.rename(columns = {"Ratings": n}, inplace = True)
    #print(column)
    table_ratings = table_ratings.join(column)
table_ratings
```

```python
table_Novelty = pd.DataFrame([df[df.Type == 0].Novelty]).T.reset_index(drop=True) 
table_Novelty.rename(columns = {"Novelty": 'Typical'}, inplace = True)
for i, n in zip(range(1,3), ['Atyplical', "Ripple"]):
    column = pd.DataFrame([df.loc[df.Type == i].Novelty]).T.reset_index(drop=True)
    column.rename(columns = {"Novelty": n}, inplace = True)
    #print(column)
    table_Novelty = table_Novelty.join(column)
table_Novelty
```
